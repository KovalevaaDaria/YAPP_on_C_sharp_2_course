### Загрузка данных из результатов моделирования:

```python
result_df = pd.read_csv("/Users/daria/Documents/Бауманка предметы/2 курс/3 семестр/ЯПП/С#/lab_6/Demographic/Exec/bin/Debug/net7.0/YearPopulationFile.csv")
result_df_age = pd.read_csv("/Users/daria/Documents/Бауманка предметы/2 курс/3 семестр/ЯПП/С#/lab_6/Demographic/Exec/bin/Debug/net7.0/AgePopulation.csv")
```

### График изменения общего населения по годам:

```python
years = result_df['Year']
population = result_df['Population']
plt.figure(figsize=(10, 6))
plt.plot(years, population, label='Общее население', c='#9896F1', marker='o')
plt.title('Изменение общего населения по годам')
plt.xlabel('Год')
plt.ylabel('Численность (тыс. чел.)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show() 
```

### График изменения населения мужского пола по годам:

```python
years = result_df['Year']
men_population = result_df['Men']
plt.figure(figsize=(10, 6))
plt.plot(years, men_population, label='Население мужского пола', color= "#42aaff", marker='o')
plt.title('Изменение населения мужского пола по годам')
plt.xlabel('Год')
plt.ylabel('Численность (тыс. чел.)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### График изменения населения женского пола по годам:

```python
years = result_df['Year']
women_population = result_df['Women']
plt.figure(figsize=(10, 6))
plt.plot(years, women_population, label='Население женского пола', color='#EC96F1', marker='o')
plt.title('Изменение населения женского пола по годам')
plt.xlabel('Год')
plt.ylabel('Численность (тыс. чел.)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
