```python
result_by_year = pd.read_csv('/Users/daria/Documents/Бауманка предметы/2 курс/3 семестр/ЯПП/С#/lab_6/Demographic/Exec/bin/Files/YearPopulationFile.csv')
result_by_age = pd.read_csv('/Users/daria/Documents/Бауманка предметы/2 курс/3 семестр/ЯПП/С#/lab_6/Demographic/Exec/bin/Files/AgePopulation.csv')
```

### График изменения общего населения по годам:

```python
years = result_by_year['Year']
population = result_by_year['Population']
plt.figure(figsize=(10, 6))
plt.plot(years, population, label='Общее население', c='#9896F1', marker='o')
plt.title('Изменение общего населения по годам')
plt.xlabel('Год')
plt.ylabel('Численность (тыс. чел.)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show() 
```

### График изменения населения мужского пола по годам:

```python
years = result_by_year['Year']
men_population = result_by_year['male']
plt.figure(figsize=(10, 6))
plt.plot(years, men_population, label='Население мужского пола', color= "#42aaff", marker='o')
plt.title('Изменение населения мужского пола по годам')
plt.xlabel('Год')
plt.ylabel('Численность (тыс. чел.)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### График изменения населения женского пола по годам:

```python
years = result_by_year['Year']
women_population = result_by_year['female']
plt.figure(figsize=(10, 6))
plt.plot(years, women_population, label='Население женского пола', color='#EC96F1', marker='o')
plt.title('Изменение населения женского пола по годам')
plt.xlabel('Год')
plt.ylabel('Численность (тыс. чел.)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### График изменения возрастного состава мужского пола на конец моделирования:

```python
male_row = result_by_age[result_by_age['Sex'] == 'Male']
men_population = male_row['0-18'].values[0], male_row['19-45'].values[0], male_row['45-65'].values[0], male_row['65-100'].values[0]

plt.figure(figsize=(10, 6))
categories = ['0-18', '19-45', '45-65', '65-100']
plt.bar(categories, men_population, color='#9896F1')
plt.title('Возрастной состав мужского населения на конец моделирования')
plt.xlabel('Возрастные категории')
plt.ylabel('Численность (тыс. чел.)')
plt.show()
```
